```python
df['studytime'] = df['studytime']/df['studytime'].max()
df['G1'] = df['G1']/df['G1'].max()
```

```python
df['studytime'] = df['studytime']/df['studytime'].max()
df[['studytime','absences','G1']].head()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["G1"])

# set x/y labels and plot title
plt.pyplot.xlabel("G1")
plt.pyplot.ylabel("count")
plt.pyplot.title("G1 bins")
```

```python
bins = np.linspace(min(df["G1"]), max(df["G1"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['G1-binned'] = pd.cut(df['G1'], bins, labels=group_names, include_lowest=True )
df[['G1','G1-binned']].head(10)
```

```python
df["G1-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["G1-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("G1")
plt.pyplot.ylabel("count")
plt.pyplot.title("G1 bins")
```

```python
dummy_variable_1 = pd.get_dummies(df["sex"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'F':'sex-F', 'M':'sex-M'}, inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("sex", axis = 1, inplace=True)
```
