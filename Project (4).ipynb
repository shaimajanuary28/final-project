```python
df['studytime'] = df['studytime']/df['studytime'].max()
df['G3'] = df['G3']/df['G3'].max()
```

```python
df['studytime'] = df['studytime']/df['studytime'].max()
df[['studytime','absences','G3']].head()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["G3"])

# set x/y labels and plot title
plt.pyplot.xlabel("G3")
plt.pyplot.ylabel("count")
plt.pyplot.title("G3 bins")
```

```python
bins = np.linspace(min(df["G3"]), max(df["G3"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['G3-binned'] = pd.cut(df['G3'], bins, labels=group_names, include_lowest=True )
df[['G3','G3-binned']].head(10)
```

```python
df["G3-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["G3-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("G3")
plt.pyplot.ylabel("count")
plt.pyplot.title("G3 bins")
```

```python
dummy_variable_1 = pd.get_dummies(df["sex"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'F':'sex-F', 'M':'sex-M'}, inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("sex", axis = 1, inplace=True)
```
